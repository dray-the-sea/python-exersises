Your friend has a coffee shop. They track sales data as follows:
- sales in a './data/coffee-shop-sales.csv' 
- prices and unit costs in './data/coffee-shop-prices.csv'
- each line in the sales spreadsheet represents the sale of a single item

Load the data and explore what's there. 
1. see the count of each item sold for each day.
2. most sold item
3. least sold item 

```python
# import pandas 
import pandas as pd 
```

```python
# read the files with the store data 
sales = pd.read_csv("./data/coffee-shop-sales.csv")
prices = pd.read_csv("./data/coffee-shop-prices.csv")
```

```python
# add column for number of items sold in each sale, set it to 1. 
# This information is implied, but it will be helpful to make it more epxlicit 

sales["SaleQuantity"] = 1
```

```python
# the count of each item sold for each day.
sale_item_groups = sales.groupby(['Date','Item'])
sale_item_groups.count()
```

```python
# most sold item 
item_groups = sales.groupby(['Item'])
item_groups.count()
```
